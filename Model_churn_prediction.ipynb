# MODEL FOR CHURN PREDICTION

```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

# Read CSV file

```python
df=pd.read_csv("tel_churn.csv")
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python

model_dt.fit(x_train,y_train)
```

```python

DecisionTreeClassifier(max_depth=6, min_samples_leaf=8, random_state=100)
```

```python

y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
import os
import multiprocessing

# Determine the number of logical cores
logical_cores = multiprocessing.cpu_count()
os.environ['LOKY_MAX_CPU_COUNT'] = str(logical_cores)
```

```python
sm = SMOTEENN()
X_resampled1, y_resampled1 = sm.fit_resample(x,y)
```

```python
xr_train1,xr_test1,yr_train1,yr_test1=train_test_split(X_resampled1, y_resampled1,test_size=0.2)
```

```python
model_dt_smote=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt_smote.fit(xr_train1,yr_train1)
yr_predict = model_dt_smote.predict(xr_test1)
model_score_r = model_dt_smote.score(xr_test1, yr_test1)
print(model_score_r)
print(metrics.classification_report(yr_test1, yr_predict))
```

```python
print(metrics.confusion_matrix(yr_test1, yr_predict))
```
